# Notebook to process Level 1 Snow Depth Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import glob
import pylab
pylab.rcParams['figure.figsize'] = (22.0, 11.0)
import scipy
```

```python
data = pd.read_pickle(
'../../data/AMS_L2_WY_2006_2012.pcl')
```

```python
wy = data['2010-10-01':'2011-09-30']
```

```python
plt.plot(wy.index, wy.s5)
```

```python
t=0.6 # threshold value [cm]
n = len(wy)
wy_filt = wy.s5.copy() # copy sensor 5
window = 1
step = 0
q = 0

for i in xrange(window,n):
    diff = np.abs(wy.s5[i]-wy.s5[i-window])
    if diff > t:
        wy_filt[i] = np.NaN
        q = q + 1
        
        
print 'Entries Removed: '+str(q)
plt.plot(wy.index, wy_filt);
```

```python
n = len(wy)
wy_filt2 = wy_filt.copy()
window = 45
t = (window**0.1) # threshold value [cm]
q = 0
for i in xrange(window,n-window):
    ave = np.nanmean(wy_filt[i-window:i+window])
    
    if wy_filt[i] > ave+t:
        wy_filt2[i] = np.NaN
        q += 1
print 'Entries Removed: '+str(q)
plt.plot(wy.index, wy_filt2)
```

```python
wy_interp = wy_filt2.interpolate('slinear')
```

```python
plt.plot(wy.index, wy_interp)
```

```python
t=0.2 # threshold value [cm]
n = len(wy)
wy_filt3 = wy_interp.copy() # copy sensor 5
window = 1
step = 0
q = 0

for i in xrange(window,n):
    diff = np.abs(wy_interp[i]-wy_interp[i-window])
    if diff > t:
        wy_filt3[i] = np.NaN
        q = q + 1
        
        
print 'Entries Removed: '+str(q)
plt.plot(wy.index, wy_filt3);
```

```python
wy_interp2 = wy_filt3.interpolate('slinear')
```

```python
plt.plot(wy.index, wy_interp2)
```

This is about as good as it gets

```python
data[wy_interp2.index.min():wy_interp2.index.max()].s5 = wy_interp2

data = data['2005-10-01':'2012-09-30']
```
