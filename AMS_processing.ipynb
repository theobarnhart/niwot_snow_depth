```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
pylab.rcParams['figure.figsize'] = (22.0, 11.0)
import scipy
```

```python
files = glob.glob('/Volumes/hydroProjects/Data/FieldData/pingersites/AMS/data/from_Emily/AMS_*.csv')
```

```python
files
```

```python
cols = ['date','s1','s2','s3','s4','s5','s6','s7','s8','s9']
```

```python
data = pd.DataFrame()
for fl in files:
    tmp = pd.read_csv(fl,names=cols, header=False, skiprows=0)
    tmp.index = pd.DatetimeIndex(tmp.date)
    data = data.append(tmp)
```

```python
data = data.sort()
```

```python
data.to_pickle('./AMS_L2_WY_2006_2012.pcl') # save the data
```
