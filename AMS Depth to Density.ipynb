```python
strt = '2005-10-01'
nd = '2014-06-25'
```

```python
swe = pd.DataFrame({'swe':niwot[strt:nd].swe,'sd':nwsd[strt:nd].sd}) # join niwot swe and niwot snow depth
swe.index = pd.Index(swe.index,freq=swe.index.freqstr)

# remove unrealistic values
swe.swe[swe.swe <= 0] = np.NaN
swe.sd[swe.sd <= 0] = np.NaN

swe['dr'] = swe.swe/swe.sd # calculate the density ratio

swe['rho_s'] = swe['dr']*rho_w
swe.rho_s[swe.rho_s > rho_w] = np.NaN
```
