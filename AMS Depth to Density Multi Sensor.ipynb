```python
strt = '2005-10-01'
nd = '2014-06-25'
```

```python
swe = pd.DataFrame({'swe':niwot[strt:nd].swe,'sd':nwsd[strt:nd].sd})
swe.index = pd.Index(swe.index,freq=swe.index.freqstr) # this is some pandas mumbo jumbo to fix a compatability problem with old pickles...
swe.swe /= 100. # convert swe from cm to m
swe.swe[swe.swe <= 0] = np.NaN
swe.sd[swe.sd <= 0] = np.NaN
# convert SWE to the weight of frozen water

swe['mw'] = swe['swe'] * rho_i

# convert sd from cm to m

swe['sd'] /= 100.

# divide the mass of water by the depth to get the density

swe['rho_s'] = swe['mw']/swe['sd']
swe.loc[swe.rho_s > rho_i,'rho_s'] = np.NaN
swe.rho_s[swe.rho_s < 0] = np.NaN
```

```python
for wyear in xrange(2006,2015):
    year=wyear-1

    strt = str(year)+'-10-01'
    nd = str(wyear)+'-09-30'


    plt.plot(swe[strt:nd].index,swe[strt:nd].rho_s);
    plt.xlim([strt,nd])
    plt.title('WY '+str(wyear), fontsize=18)
    plt.show()
    
```

```python
for wyear in xrange(2006,2015):
    year=wyear-1

    strt = str(year)+'-12-01'
    nd = str(wyear)+'-09-30'
    
    dat = swe[strt:nd] # extract the data
    
    mx_rho = str(dat.rho_s.idxmax()) # grab the date of maximum density
    
    dat[mx_rho:nd] = rho_i # fill from the date of maximum density to the end of the water year 
    
    dat.rho_s = dat.rho_s.interpolate(how='linear') # interpolate the holes
    
    swe[strt:nd]=dat[strt:nd] # reinsert the data
```

```python
for wyear in xrange(2006,2015):
    year=wyear-1

    strt = str(year)+'-10-01'
    nd = str(wyear)+'-09-30'


    plt.plot(swe[strt:nd].index,swe[strt:nd].rho_s);
    plt.xlim([strt,nd])
    plt.title('WY '+str(wyear), fontsize=18)
    plt.show()
```
