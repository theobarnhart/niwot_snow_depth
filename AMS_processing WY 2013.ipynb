```python
cols = ['tmp','Year','DOY','Time','Batt_Volt_Min','DSTempC_1','DSTempF_1','Depth1','DSTempC_2','DSTempF_2',
        'Depth2','DSTempC_3','DSTempF_3','Depth3','DSTempC_4','DSTempF_4','Depth4','DSTempC_5','DSTempF_5','Depth5',
        'DSTempC_6','DSTempF_6','Depth6','DSTempC_7','DSTempF_7','Depth7','DSTempC_8','DSTempF_8','Depth8','DSTempC_9',
        'DSTempF_9','Depth9','DSTempC_10','DSTempF_10','Depth10']
```

```python
def dtparse(x,y,z):
    # z = hours
    # x = year
    # y = day of year
    
    if len(str(z)) == 3:
        z = '0'+z # add a leading zero
        h = int(z[0:2]) # pull out hour
        m = int(z[2:4]) # pull out minutes
    else:
        h = int(z[0:2]) # pull out hour
        m = int(z[2:4]) # pull out minutes
    if h == 24 and y < 366:
        y = int(y) + 1
        h = 0
    if h == 24:
        h=0
    
        
        
    return pd.datetime.strptime(str(x)+' '+str(y)+' '+str(h)+' '+str(m),'%Y %j %H %M')
```

def parser(m,d,y,hhmm):
    if len(hhmm) < 4:
        hhmm = int('0'+str(hhmm))
    hh = hhmm[0:2]
    mm = hhmm[2:4]
    
    
    
    date = str(m)+"/"+str(d)+"/"+str(y)+" "+str(hh)+":"+str(mm)
    return pd.datetime.strptime(date, '%m/%d/%Y %H:%M')

```python
data = pd.DataFrame()
for fl in files:
    tmp = pd.read_csv(fl,names=cols, header=False, skiprows=0, parse_dates={'date':[1,2,3]}, date_parser=dtparse)
    tmp.index = pd.DatetimeIndex(tmp.date)
    data = data.append(tmp)
```

```python
data = data.sort()
```

```python
sd = data.filter(regex='Depth')
```

# Clean the noise from the data:

```python
sdc = pd.DataFrame(np.zeros_like(sd),columns=sd.columns,index=sd.index)

sdc.head()

wrt = True
```

## Sensor 1

```python
sens = 1

dat = pd.DataFrame({'depth': sd['Depth'+str(sens)].copy()},index=sd.index)
plt.subplot(211)
plt.plot(dat.index, dat.depth);
plt.ylim([-10,250])
plt.axhline(0)
plt.title('Sensor '+str(sens)+' raw', fontsize=16)


dat[dat.depth >200] = np.NaN

dat2 = dat.resample('2H', how = np.mean)
diff = np.abs(dat-dat2)
dat[diff>0.01] = np.NaN

dat = dat.interpolate(how='spline',limit=48)

ct = 0
for i in range(1,len(dat)):
    if np.abs(dat.ix[i].depth-dat.ix[i-1].depth) > 0.0001:
        dat.ix[i].depth = np.NaN
        ct +=1
dat = dat.interpolate(how='spline',limit=48)

for i in range(0,len(dat)-1):
    if np.abs(dat.ix[i].depth - dat.ix[i+1].depth) > 0.1:
        dat.ix[i].depth = np.NaN
        ct+=1



dat = dat.interpolate(how='spline',limit=48)
plt.subplot(212)
plt.plot(dat.index, dat.depth,'r-');
plt.ylim([-10,250])
plt.axhline(0);
plt.title('Sensor '+str(sens)+' clean', fontsize=16);

if wrt == True:
    sdc['Depth'+str(sens)] = dat.depth
```
