```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import glob
import pylab
pylab.rcParams['figure.figsize'] = (22.0, 11.0)
import scipy
```

```python
dat = pd.read_csv('../../greenness_timing/snow_depth/niwot_snotel_manual.csv',skiprows=201)
dat.index = pd.DatetimeIndex(dat.date)
del dat['date']
```

```python
dat
```

```python
# convert from in to cm
dat *= 2.54
```

```python
# convert to m
dat /= 100
```

```python
plt.plot(dat.index,dat.swe)
```

```python
# calculate the density of the snow

dat['dr'] = dat.swe/dat.mean_depth
dat['rho_s']=dat.dr*1000.
```

```python
plt.plot(dat.index,dat.rho_s)
```
