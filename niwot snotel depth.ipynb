```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
pylab.rcParams['figure.figsize'] = (16.0, 8.0)
import scipy
```

```python
files = glob.glob('../../data/niwot_sd/*.csv')
```

```python
files
```

```python
def dtparse(date,time):
    spc = ' '
    date_time = ''.join([date,spc,time])
    return pd.datetime.strptime(date_time,'%Y-%m-%d %H:%M')
```

```python
data = pd.DataFrame()

for fl in files:

    wyear = int(fl.split('=')[-1][0:4])
    year = wyear-1 
    dat = pd.read_csv(fl,skiprows=2, date_parser=dtparse, parse_dates={'date_time':[1,2]})
    dat.index = pd.DatetimeIndex(dat.date_time)
    del dat['date_time']
    del dat['Site Id']
    del dat['Unnamed: 4']
    data = data.append(dat)

data.columns = ['sd']
```

```python
data.sd[data.sd > 75] = np.NaN
data.sd[data.sd < -10] = np.NaN

data = data.interpolate(how = 'spline', limit=48)

plt.plot(data.index,data.sd);
```

```python
strt = '2014-7-01'
nd = '2014-8-15'

data[strt:nd].sd[data.sd > 0] = np.NaN

plt.plot(data[strt:nd].index,data[strt:nd].sd)
```

```python
plt.plot(data.index,data.sd);
```

```python
# convert snow depth from in to cm

data.sd *= 2.54
```

```python
data2 = data.resample('D',how=np.nanmean)

plt.plot(data2.index, data2.sd)
```
