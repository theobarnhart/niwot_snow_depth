# Clean the Data

Not going for the best time series, but one that is plausible. 

```python
sdc = pd.DataFrame(np.zeros_like(sd),columns=sd.columns,index=sd.index)

sdc.head()
```

## Sensor 1

```python
dat = pd.DataFrame({'depth' : sd['Depth1'].copy()})
dat = dat.sort()
dat[dat.depth > 200] = np.NaN
dat[dat.depth < -10] = np.NaN

dat2 = dat.resample('2H', how = np.mean)
diff = np.abs(dat-dat2)
dat[diff>0.01] = np.NaN

dat = dat.interpolate(how='spline',limit=48)

#dat2 = dat.resample('48H', how = np.max)
#diff = np.abs(dat-dat2)
#dat[diff>0.04] = np.NaN
ct = 0
for i in range(1,len(dat)):
    if np.abs(dat.ix[i].depth-dat.ix[i-1].depth) > 0.0001:
        dat.ix[i].depth = np.NaN
        ct +=1
dat = dat.interpolate(how='spline',limit=48)

for i in range(0,len(dat)-1):
    if np.abs(dat.ix[i].depth - dat.ix[i+1].depth) > 0.1:
        dat.ix[i].depth = np.NaN
        ct+=1



dat = dat.interpolate(how='spline',limit=48)
plt.plot(dat.index, dat.depth, 'c-')
plt.axhline(0)
print "Values Removed: "+str(ct)
sdc.Depth1=dat.depth
```
